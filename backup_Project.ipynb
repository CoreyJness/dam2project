Now, import necessary modules and load the data into a variable that can used to create a data frame.

```python
import pandas as pd
import numpy as np
housing_prices = pd.read_csv('housing_prices.csv')
loan_data = pd.read_csv('loan_data.csv')
```

Concat the two datasets into one large data frame:

```python
mortgage_planning_data = pd.concat([housing_prices, loan_data], axis=1, ignore_index=False)
```
